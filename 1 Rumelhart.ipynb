## Forward propagation through MLP with 2 intermediate layers
Our input and output layers both have 8 nodes, while the intermediate layers have 64 nodes.
We use a sigmoid activation function.

```python
np.random.seed(42)

input = np.random.randn(1, 8)
w = [np.random.randn(8, 64), np.random.randn(64, 64), np.random.randn(64, 8)]
b = [np.random.randn(1, 64), np.random.randn(1, 64), np.random.randn(1, 8)]

def sigmoid(input: np.ndarray):
    return 1 / (1 + np.exp(-input))

for i in range(len(w)):
    print(input)
    input = sigmoid(np.dot(input, w[i]) + b[i])
print(input)
```

Representing the same neural network using heatmaps looks pretty cool:

```python
input = np.random.randn(1, 8)
w = [np.random.randn(8, 64), np.random.randn(64, 64), np.random.randn(64, 8)]
b = [np.random.randn(1, 64), np.random.randn(1, 64), np.random.randn(1, 8)]

def sigmoid(x: np.ndarray):
    return 1 / (1 + np.exp(-x))

def plot_all_layers(layers):

    plt.figure(figsize=(12, len(layers) * 3))
    
    for i, layer_values in enumerate(layers):
        plt.subplot(len(layers), 1, i + 1)
        if layer_values.shape == (1, 8):  
            plt.imshow(layer_values, cmap='Blues', aspect='auto')
        else:  
            plt.imshow(layer_values, cmap='Blues', aspect='auto')
        plt.colorbar(label='Activation Value')
        plt.title(f'Layer {i}')
    
    plt.tight_layout()  
    plt.show()

all_layers = []

for i in range(len(w)):

    pre_activation = np.dot(input, w[i]) + b[i]
    input = sigmoid(pre_activation)
    all_layers.append(input)

plot_all_layers([np.random.randn(1, 8)] + all_layers) 
```
